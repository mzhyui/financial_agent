```python
import os
import pandas as pd

df = pd.read_csv('/home/jovyan/output/agentA/train.csv')
# Print basic information about the dataset
print("Dataset Shape:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\n" + "="*50)

# Display first few rows
print("First 5 rows:")
print(df.head())
print("\n" + "="*50)

# Display basic info about the dataframe
print("Dataset Info:")
print(df.info())
print("\n" + "="*50)

# Display statistical summary
print("Statistical Summary:")
print(df.describe())
print("\n" + "="*50)

# Check for missing values
print("Missing Values:")
print(df.isnull().sum())
print("\n" + "="*50)

# Display column names and data types
print("Columns and Data Types:")
for col in df.columns:
    print(f"{col}: {df[col].dtype}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numeric columns found: {len(numeric_columns)}")
print("Numeric columns:", numeric_columns[:10], "...")  # Show first 10

# Calculate correlation matrix for numeric columns only
correlation_matrix = df[numeric_columns].corr()

# Get correlations with the target label
label_correlations = correlation_matrix['label'].sort_values(ascending=False)
print("\nCorrelation with Label (sorted):")
print(label_correlations)
```
