```python
import numpy as np

# Estimate the value of pi using the Monte Carlo method
def monte_carlo_pi(num_samples: int = 100000) -> float:
    x = np.random.uniform(0, 1, num_samples)
    y = np.random.uniform(0, 1, num_samples)
    inside_circle = (x**2 + y**2) <= 1
    pi_estimate = 4 * np.sum(inside_circle) / num_samples
    return pi_estimate

# Example usage
pi_approx = monte_carlo_pi(100000)
print(f"Estimated value of pi: {pi_approx}")
```

```python
import matplotlib.pyplot as plt

num_samples = 10000
x = np.random.uniform(0, 1, num_samples)
y = np.random.uniform(0, 1, num_samples)
inside_circle = (x**2 + y**2) <= 1

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(x[inside_circle], y[inside_circle], color='blue', s=1, label='Inside Circle')
ax.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1, label='Outside Circle')
circle = plt.Circle((0, 0), 1, color='black', fill=False, linewidth=2)
ax.add_artist(circle)
ax.set_aspect('equal')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title(f"Monte Carlo Pi Estimation\nEstimated   {4 * np.sum(inside_circle) / num_samples:.5f}")
ax.legend()
plt.show()
```
